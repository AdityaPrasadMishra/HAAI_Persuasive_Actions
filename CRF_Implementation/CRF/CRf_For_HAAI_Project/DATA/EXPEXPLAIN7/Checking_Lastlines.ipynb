```python
import numpy as np
```

```python
#Finding max count to 
arrays = []
maxcount = 0
with open("aditya_setworking") as file:
    for line in file:
        new_array =[i.strip('\n') for i in line.split(' ') if i != '']
        if len(new_array) > maxcount:
            maxcount = len(new_array)
        arrays.append(new_array)
```

```python
newarray = []
for arr in arrays:
    arrw = []
    for r in range(maxcount-1):
        if r >=len(arr)-1:
            arrw.append("blank")            
        else:
            arrw.append(arr[r])
            
    arrw.append(arr[len(arr)-1])
    newarray.append(arrw)
newarray
```

```python
dictnumcoll = {}
ity = 1
numericnumarray =[]
for arr in newarray:
    numarray = []
    for evr in arr:
        if evr not in dictnumcoll:
            dictnumcoll[evr] = ity
            numarray.append(ity)
            ity+=1
        else:
            numarray.append(dictnumcoll[evr])
    numarray2 = sorted(numarray[0:(len(numarray)-1)])
    numarray2.append(numarray[len(numarray)-1])
    numericnumarray.append(numarray2)
```

```python
numericnumarray = np.array(numericnumarray)
```

```python
X = numericnumarray[:, :-1]
Y = numericnumarray[:, -1]
X
Y
```

```python
from sklearn import datasets
iris = datasets.load_iris()
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
y_pred = gnb.fit(X,Y).predict(X)
print("Number of mislabeled points out of a total %d points : %d"% (X.shape[0],(Y != y_pred).sum()))
gnb.score(X,Y)
```

```python
from sklearn.svm import SVC
clf = SVC(gamma='auto')
y_pred = clf.fit(X, Y).predict(X)
print("Number of mislabeled points out of a total %d points : %d"% (X.shape[0],(Y != y_pred).sum()))
```

```python
clf.score(X,Y)
```

```python
from sklearn.naive_bayes import MultinomialNB
gnb = MultinomialNB()
gnb.fit(X,Y)
gnb.score(X,Y)
```

```python
from sklearn.naive_bayes import ComplementNB
gnb = ComplementNB()
gnb.fit(X,Y)
gnb.score(X,Y)
```
